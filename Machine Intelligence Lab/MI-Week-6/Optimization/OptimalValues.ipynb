```python
import pandas as pd
import numpy as np
```

```python
runs = ['./run1.txt',
'./run2.txt',
'./run3.txt',
'./run4.txt',
'./run5.txt']
```

```python
def parse(file):
    df = pd.read_csv(file,delimiter=',',header=None)
    df.columns = ['kernel','C','Gamma','train_acc','test_acc']
    df['kernel'] = 'rbf'
    c = df['C'].str.split(" ",expand=True)
    gamma = df['Gamma'].str.split(" ",expand=True)
    train = df['train_acc'].str.split(" ",expand=True)
    test = df['test_acc'].str.split(" ",expand=True)
    df['C'] = c[2]
    df['Gamma'] = gamma[2]
    df['train_acc'] = train[2]
    df['test_acc'] = test[2]
    df['C'] = df['C'].astype(float)
    df['Gamma'] = df['Gamma'].astype(float)
    df['train_acc'] = df['train_acc'].astype(float)
    df['test_acc'] = df['test_acc'].astype(float)
    return df
```

```python
dfs = []
for run in runs:
    dfs.append(parse(run))
df = pd.concat(dfs)
df = df.reset_index(drop=True)
```
