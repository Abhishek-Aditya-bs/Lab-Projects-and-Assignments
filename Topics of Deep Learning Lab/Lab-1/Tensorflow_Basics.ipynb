```python
import tensorflow as tf

print(tf.__version__)
```

 # Creating a Tensor

 Rank - 1 Tensor
 Constant Tensors are Immuatable

```python
rank1_tensor = tf.constant([0,1,43,2,3,4,8,13,21])
print(rank1_tensor)
```

# Representing Scalar

Scalar can be represented as rank-0 Tensor

```python
scalar = tf.constant(5)
print(scalar)
```

# Rank-2 Tensors (nested lists)

```python
rank2_tensor = tf.constant([[1,2,3], [4,5,6]])
print(rank2_tensor)
```

# Defining with Type

By default type is inferred by tensorflow based on input

### Supported Types

- float16, float32, float64
- int8, int16, int32, int64
- uint8, uint16 (unsigned)
- string
- bool
- complex64, complex128
- qint8, qint32(quantized)

```python
rank2_tensorfloat = tf.constant([[1,2], [3,4], [5,6]], dtype=tf.float32)
print(rank2_tensorfloat)
```

# Attributes / Properties of Tensors

```python
print(f'Tensor Shape : {rank2_tensor.shape}')
print(f'Tensor Rank : {rank2_tensor.ndim}')
print(f'Type of all elements : {rank2_tensor.dtype}')
print(f'Total Number of elements : {tf.size(rank2_tensor.shape)}')
```

# Converting Numpy array to Tensor

```python
np_arr = np.array([[1,2,3],[4,5,6]])
tensor_frm_np = tf.constant(np_arr)
print(tensor_frm_np)
```

# Indexing Tensors

```python
print('Tensor :', rank1_tensor)
print('Index 2 :', rank1_tensor[2])
print('Index 4 :', rank1_tensor[4])
```

# Slicing Tensors

```python
print('Slicing 3-6 (including 3, excluding 6) : ', rank1_tensor[3:6])
print('Slicing with step(2) : ', rank1_tensor[3:8:2])
print('All elements : ', rank1_tensor[:])
print('From Start of Tensor : ', rank1_tensor[:4])
print('To end of the tensor : ', rank1_tensor[4:])
```

# Multi Indexing Tensors

```python
print('Tensor :' , rank2_tensor)
print('Specific element: ', rank2_tensor[1][2])
print('Specific elumunt with "," operator: ', rank2_tensor[1, 2])
print('Slicing across different axes : ', rank2_tensor[1:, :])
```
