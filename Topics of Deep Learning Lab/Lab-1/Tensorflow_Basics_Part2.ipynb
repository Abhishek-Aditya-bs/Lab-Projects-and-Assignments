## Defining a Variable in Tensorflow

```python
a = tf.Variable(0,name="input")
b = tf.constant(1)
mid = a + b
update_value = a.assign(mid)
tg = tf.global_variables_initializer()

with tf.Session() as sess:
  sess.run(tg)
  print(sess.run(a))

  for i in range(3):
    sess.run(update_value)
    print(sess.run(a))
```

## Gradient Descent Algorithm

$ x := x - ηδx $

$ η $ is the learning rate 

$ δx $ is the gradient of loss with respect to $x$ 

```python
import numpy as np

inp = 3
out = -inp

w = 0.5

i,o,w = [tf.convert_to_tensor(float(x)) for x in [inp,out,w]]

for it in range(40):
  with tf.GradientTape() as tape:
    tape.watch(w)
    loss = (o - (i*w))**2
  grad = tape.gradient(loss,w)
  w += -grad * 0.01
  print(f"Loss :{loss.eval(session=tf.Session())} Weight : {w.eval(session=tf.Session())}")
```
